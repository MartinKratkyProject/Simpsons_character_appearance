```python
import requests
from bs4 import BeautifulSoup
chars = []
def crawl(url, target_id):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        target_element = soup.find(id=target_id)
        
        if target_element:
            for link in target_element.find_all('a'):
                href = link.get('title')
                if href:
                    chars.append(href)
        else:
            print(f"No element found with the ID: {target_id}", url)

    else:
        print(f"Failed to retrieve page: {response.status_code}", url)
```

```python
pole = []
for i in df['Episode']:
    pole.append(i)
pole
```

```python
for i in df.index:
    url = ('https://simpsons.fandom.com/wiki/'+pole[i]+'/Appearances')
    crawl(url, 'gallery-0')
```
