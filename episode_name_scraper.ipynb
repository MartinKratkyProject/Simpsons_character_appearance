```python
import requests
from bs4 import BeautifulSoup

pole = []
def crawl(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        for link in soup.find_all('a'):
            print(link.get('href'))
            pole.append(link.get('href'))
    else:
        print(f"Failed to retrieve page: {response.status_code}")
crawl('https://simpsons.fandom.com/wiki/List_of_Episodes')
```

```python
import pandas as pd
```

```python
for i in df.index:
     if (df.at[i, 0][:5]=='https'):
          df.drop(axis=0, index=i, inplace=True)
     else:
          df.at[i, 0] = df.at[i, 0][6:]
```

```python
df
```

```python
df.replace('%27', "'", inplace=True, regex=True)
df.replace('%26', "&", inplace=True, regex=True)
df.replace('%22', '"', inplace=True, regex=True)
df.replace("%24", '$', inplace=True, regex=True)
df.replace("%D0%AF", '', inplace=True, regex=True)
df
```

```python
df.rename(columns={0: 'Episode'}, inplace=True)
```

```python
df.head(20)
```

```python
season_col = []
s = 0
for i in df.index:
    if (df.at[i, 'Episode'][:7] == 'Season_'):
        s += 1
        season_col.append(s)
    else:
        season_col.append(s)
```

```python
df["Season"] = season_col
```

```python
for i in df.index:
    if (df.at[i, 'Episode'][:7] =='Season_'):
        df.drop(index=i, inplace=True)
    else:
        if (df.at[i, 'Episode'] =='The Simpsons Movie'):
            df.drop(index=i, inplace=True)
```
